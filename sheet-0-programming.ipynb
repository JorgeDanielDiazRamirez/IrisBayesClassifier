### Creating $x$ and $y$

```python
y = np.linspace(3, 3*n, n)
print(y)
```

```python
y = np.array(range(3, 3*(n+1), 3), float)
print(y)
```

### Creating $y_2$

```python
y2 = np.round(15 * np.sin(y))
print(y2)
```

```python
msin = lambda t: np.round(15 * np.sin(t)) # This is a vectorized solution, maybe a bit overkill at this point
vfunc = np.vectorize(msin)
y2 = vfunc(y)
print(y2)
```

```python
np.where(y2 > 6)
```

```python
np.all(y2 > 2)
```

```python
np.count_nonzero(y2 == 5)
```

## Exercise 2.2

```python
M = np.array([y, y2])
print(M)
```

```python
## First way to create the column based matrix
M2a = np.transpose(M)
print(M2a)
```

```python
# Second possibility
M2b = np.column_stack([y,y2])
print(M2b)
```

```python
# Third possibility
M2c = np.reshape(y, (n,1))
M2c = np.append(M2c, np.reshape(y2, (n,1)), axis=1)
print(M2c)
```

```python
# Check that the 3 matrices are in fact equal
print (np.all(M2a == M2b))
print (np.all(M2b == M2c))
```

```python
import matplotlib.pyplot as plt
import itertools as it
```

```python
L = ["a", "b", "c"]
L1 = [s + "x" for s in L]
print(L1)
```

```python
# Plot using the group function
groups = df.groupby('species') ## this is like grouping in SQL
fig, ax = plt.subplots(3, 2, figsize=(10, 14))
att_combos = [(df.columns[i[0]], df.columns[i[1]]) for i in it.combinations(range(4), 2)]
for i, combo in enumerate(att_combos):
    row = i % 3
    col = int(np.floor(i / 3))
    a = ax[row][col]
    for name, group in groups:
        a.scatter(group[combo[0]], group[combo[1]], marker='o', label=name)
    a.set_xlabel(combo[0])
    a.set_ylabel(combo[1])
    a.legend()
fig.savefig("plot.pdf")
```

```python
boxes = []
labels = []
for col in df.columns:
    if col != "species":
        boxes.append(df[col].values)
        labels.append(col)
fig, ax = plt.subplots()
ax.boxplot(boxes)
ax.set_xticklabels(labels)
plt.show()
```
